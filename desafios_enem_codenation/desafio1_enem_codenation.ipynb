```python
# Fazendo uma copia do objeto "df(que contem todas os registro do arquivo csv)" porem, copiando apenas as colunas que nos interessa.
df_notas = df[["NU_INSCRICAO", 
               "NU_NOTA_CN", 
               "NU_NOTA_CH", 
               "NU_NOTA_LC", 
               "NU_NOTA_MT", 
               "NU_NOTA_REDACAO"]]
```

```python
# Mostrando as 10 primeiras linha do novo objeto. 
df_notas.head(10)
```

```python
# Mostrando as todas as informaçoes contidas no arquivo csv. As informaçoes que podemos ver são as seguintes: 
# Nomes das colunas | Qtd. de registros não nulos | e o tipo de dado contindo em cada coluna
df_notas.info()
```

```python
# Primeira forma usada para o calculo da nota final
df_notas["NOTA_FINAL"] = (
    df_notas["NU_NOTA_MT"] * 3 + 
    df_notas["NU_NOTA_CN"] * 2 + 
    df_notas["NU_NOTA_CH"] * 1 +
    df_notas["NU_NOTA_LC"] * 1.5 + 
    df_notas["NU_NOTA_REDACAO"] * 3) / (3 + 2 + 1 + 1.5 + 3)
```

```python
df_notas.head()
```

```python
# Segunda forma HARDCORE feita pelo Vinicius
import numpy as np  # Importa  a biblioteca NUmpy
w = np.array((3,2,1,1.5,3)) # Criando uma tuple com os peso das notas
w = w/sum(w) # Vinicius por favor exlicar o que acontece aqui 8^)
w = w.T # E aqui tambem ;)

# Calculando as nota finais
df_notas['NOTA_FINAL_ALT'] = np.dot(df_notas[["NU_NOTA_MT","NU_NOTA_CN","NU_NOTA_CH","NU_NOTA_LC","NU_NOTA_REDACAO"]].values,w)
```

```python
df_notas.head()
```

```python
# Pegando o 20 primeiros registros decrecente das notas e salvando em um aquivo csv
df_notas[["NU_INSCRICAO","NOTA_FINAL"]].sort_values(
    by=["NOTA_FINAL"], ascending=False
).head(20).to_csv("answer.csv",index=False)
```
